```python
from pybaseball import statcast
def fetch_statcast_for_season(season, chunk_size_days=30):
    start = pd.to_datetime(f"{season}-03-01")
    end = pd.to_datetime(f"{season}-11-01")
    all_dfs = []
    cur = start
    while cur < end:
        chunk_end = min(cur + pd.Timedelta(days=chunk_size_days - 1), end)
        print(f"Fetching {cur.date()} to {chunk_end.date()} …")
        df = statcast(start_dt=cur.strftime("%Y-%m-%d"), end_dt=chunk_end.strftime("%Y-%m-%d"))
        all_dfs.append(df)
        cur = chunk_end + pd.Timedelta(days=1)
    full = pd.concat(all_dfs, ignore_index=True)
    return full

df2025 = fetch_statcast_for_season(2025)
swings = df2025[df2025['type'] == 'X']
swings.to_parquet("2025_swings.parquet")
```
